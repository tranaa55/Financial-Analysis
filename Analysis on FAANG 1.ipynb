## Daily Percent Change


Daily Percent Change analysis is used to measure the stocks' volatility. If daily return have a wide distribution, the stock is more volatile from one day to the next.

```python
#stock['Daily Return'] = (stock['Close']/stock['Close'].shift(1))-1
fb['Daily Return'] = fb['Close'].pct_change(1)
amzn['Daily Return'] = amzn['Close'].pct_change(1)
aapl['Daily Return'] = aapl['Close'].pct_change(1)
nflx['Daily Return'] = nflx['Close'].pct_change(1)
goog['Daily Return'] = goog['Close'].pct_change(1)
```

```python
fb['Daily Return'].hist(figsize = (16,8),label = 'FB',bins = 100, alpha = .3)
amzn['Daily Return'].hist(label = 'Amazon', bins = 100,alpha = .3)
aapl['Daily Return'].hist(label = 'Apple',bins = 100, alpha = .3)
nflx['Daily Return'].hist(label = 'Netflix',bins = 100, alpha =.3)
goog['Daily Return'].hist(label = 'Google',bins = 100, alpha = .5)
plt.legend()
```

```python
fb['Daily Return'].plot.density(figsize = (16,8),label = 'FB')
amzn['Daily Return'].plot.density(label = 'Amazon')
aapl['Daily Return'].plot.density(label = 'Apple')
nflx['Daily Return'].plot.density(label = 'Netflix')
goog['Daily Return'].plot.density(label = 'Google')
plt.legend()
```

Netflix displayed a wider distribution demonstrating a higher stock volatility than its tech counterpart

## Cumulative Return

Daily Return is useful to measure volatility, but it does not give an investor an insight into the gains he had made till date. Cumulative return is computed relative to the day investment is made. If cumulative return is above 1, you are making profits.

The formula for a cumulative daily return is:

df[daily_cumulative_return] = ( 1 + df[pct_daily_return] ).cumprod()



```python
#cumprod method is used 
fb['Cumulative Return'] = (1 + fb['Daily Return']).cumprod()
amzn['Cumulative Return'] = (1 + amzn['Daily Return']).cumprod()
aapl['Cumulative Return'] = (1 + aapl['Daily Return']).cumprod()
nflx['Cumulative Return'] = (1 + nflx['Daily Return']).cumprod()
goog['Cumulative Return'] = (1 + goog['Daily Return']).cumprod()
```

```python
fb['Cumulative Return'].plot(figsize = (12,8),label = 'FB')
amzn['Cumulative Return'].plot(label = 'AMZN')
aapl['Cumulative Return'].plot(label='AAPL')
nflx['Cumulative Return'].plot(label = 'NFLX')
goog['Cumulative Return'].plot(label = 'GOOG')
plt.legend()
```
